# Sync Local Project To Remote
Use the next two cells to copy the local `Lab0` project into the remote Jupyter environment without SSH.

Workflow:
1. On your local machine, create `Lab0.zip`, base64-encode it, and copy it to the clipboard.
2. Paste the base64 text into the `payload` string in the next cell.
3. Run the sync cell, then run the verification cell.

```python
import base64
import shutil
import zipfile
from pathlib import Path

payload = """PASTE_BASE64_HERE"""

target_root = Path("/Lab0")
tmp_zip = Path("/tmp/lab0_upload.zip")
extract_root = Path("/tmp/lab0_extract")
skip_names = {".git", ".mypy_cache", "__pycache__", ".ipynb_checkpoints"}

if payload.strip() == "PASTE_BASE64_HERE":
    raise ValueError("Paste the base64 string generated on your local machine into payload first.")

if extract_root.exists():
    shutil.rmtree(extract_root)
extract_root.mkdir(parents=True, exist_ok=True)

tmp_zip.write_bytes(base64.b64decode(payload))

with zipfile.ZipFile(tmp_zip, "r") as archive:
    archive.extractall(extract_root)

source_root = extract_root / "Lab0"
if not source_root.exists():
    raise FileNotFoundError(f"Expected {source_root} inside the uploaded archive")

for path in source_root.rglob("*"):
    rel = path.relative_to(source_root)
    if any(part in skip_names for part in rel.parts):
        continue

    dest = target_root / rel
    if path.is_dir():
        dest.mkdir(parents=True, exist_ok=True)
    else:
        dest.parent.mkdir(parents=True, exist_ok=True)
        shutil.copy2(path, dest)

print(f"Synced project to {target_root}")
print("You can now import modules from /Lab0")
```
