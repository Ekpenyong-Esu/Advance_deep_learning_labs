## Section 2 — Task 0.2.1: AlexNet Transfer Learning on CIFAR-10

AlexNet was pretrained on **ImageNet** (1,000 classes, 224×224 images).  
We adapt it for **CIFAR-10** (10 classes) two different ways:

- **Fine-tuning** — all layers are updated → higher accuracy, more compute  
- **Feature extraction** — backbone is frozen, only the final layer trains → faster, slightly lower accuracy

In both cases the original last layer `Linear(4096 → 1000)` is replaced with `Linear(4096 → 10)`.
