```python
from wandb_logger import WandbLogger
import wandb
wandb.login()
```

```python
logger = WandbLogger(
    project="lab-0",
    run_name="notebook_run_test",
    config={
        "epochs": 5,
        "lr": 0.001
    },
    mode="online"   # or "offline"
)
```

```python
import random

for epoch in range(5):
    loss = random.uniform(0.5, 1.5)
    acc = random.uniform(50, 100)

    logger.log_metrics(
        epoch=epoch,
        loss=loss,
        accuracy=acc
    )

    print(f"Epoch {epoch}: loss={loss:.3f}, acc={acc:.2f}")
```

```python
logger.finish()
```
